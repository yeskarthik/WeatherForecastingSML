```python
df = pd.read_csv('dataset/SEA.csv', header=0)
```

```python
df['DATE_HOUR'] = df['DATE'].apply(lambda x: x.split(':')[0])
```

```python
tf = df.groupby('DATE_HOUR')
```

```python
df['HOURLYVISIBILITY'] = df['HOURLYVISIBILITY'].apply(lambda x: str(x)[0:len(x)-1] if str(x)[-1] == 'V' or str(x)[-1] == 's' else x)
```

```python
df['HOURLYVISIBILITY'].unique()
```

```python
df['HOURLYVISIBILITY'] = pd.to_numeric(df['HOURLYVISIBILITY'], errors='coerce')
```

```python
df['HOURLYPrecip'] = df['HOURLYPrecip'].apply(lambda x: str(x)[0:len(x)-1] if str(x)[-1] == 'T' or str(x)[-1] == 's' else x)
```

```python
df['HOURLYPrecip'] = pd.to_numeric(df['HOURLYPrecip'], errors='coerce')
```

```python
df['HOURLYDewPointTempF'] = df['HOURLYDewPointTempF'].apply(lambda x: str(x)[0:len(x)-1] if str(x)[-1] == 'T' or str(x)[-1] == 's' else x)
```

```python
df['HOURLYDewPointTempF'] = pd.to_numeric(df['HOURLYDewPointTempF'], errors='coerce')
```

```python
df['HOURLYWindSpeed'] = pd.to_numeric(df['HOURLYWindSpeed'], errors='coerce')
```

```python
df['HOURLYWindDirection'] = df['HOURLYWindDirection'].apply(lambda x: "" if str(x)[-1] == 'B' else x)
df['HOURLYWindDirection'] = pd.to_numeric(df['HOURLYWindDirection'], errors='coerce')
```

```python
df['HOURLYStationPressure'] = df['HOURLYStationPressure'].apply(lambda x: "" if str(x)[-1] == 's' else x)
df['HOURLYStationPressure'] = pd.to_numeric(df['HOURLYStationPressure'], errors='coerce')
```

```python
df['HOURLYSeaLevelPressure'] = df['HOURLYSeaLevelPressure'].apply(lambda x: "" if str(x)[-1] == 's' else x)
df['HOURLYSeaLevelPressure'] = pd.to_numeric(df['HOURLYSeaLevelPressure'], errors='coerce')
```

```python
df['HOURLYAltimeterSetting'] = df['HOURLYAltimeterSetting'].apply(lambda x: "" if str(x)[-1] == 's' else x)
df['HOURLYAltimeterSetting'] = pd.to_numeric(df['HOURLYAltimeterSetting'], errors='coerce')
```

```python
def compute_metrics(row):
    #print "HV",row['DATE_HOUR'],"mean:", row['HOURLYVISIBILITY'].mean(skipna=True)
    result = {
        'HOURLYSKYCONDITIONS': row['HOURLYSKYCONDITIONS'].str.cat(sep='::'),
        'HOURLYVISIBILITY': row['HOURLYVISIBILITY'].mean(skipna=True),
        'HOURLYPRESENTWEATHERTYPE': row['HOURLYPRSENTWEATHERTYPE'].str.cat(sep='::'),
        'HOURLYDRYBULBTEMPF':row['HOURLYDRYBULBTEMPF'].mean(skipna=True),
        'HOURLYWETBULBTEMPF':row['HOURLYWETBULBTEMPF'].mean(skipna=True),
        'HOURLYDEWPOINTTEMPF':row['HOURLYDewPointTempF'].mean(skipna=True),
        'HOURLYRELATIVEHUMIDITY':row['HOURLYRelativeHumidity'].mean(skipna=True),
        'HOURLYWINDDIRECTION': row['HOURLYWindDirection'].mean(skipna=True),
        'HOURLYWINDSPEED': row['HOURLYWindSpeed'].mean(skipna=True),
        'HOURLYWINDGUSTSPEED': row['HOURLYWindGustSpeed'].mean(skipna=True),
        'HOURLYSTATIONPRESSURE': row['HOURLYStationPressure'].mean(skipna=True),
        'HOURLYPRESSURETENDENCY': row['HOURLYPressureTendency'].median(skipna=True),
        'HOURLYPRESSURECHANGE': row['HOURLYPressureChange'].mean(skipna=True),
        'HOURLYSEALEVELPRESSURE': row['HOURLYSeaLevelPressure'].mean(skipna=True),
        'HOURLYPRECIP': row['HOURLYPrecip'].mean(skipna=True),
        'HOURLYALTIMETERSETTING': row['HOURLYAltimeterSetting'].mean(skipna=True),
        'DAILYSUNRISE': row['DAILYSunrise'].median(skipna=True),
        'DAILYSUNSET': row['DAILYSunset'].median(skipna=True)
    }
    return pd.Series(result, name='metrics')
```

```python
tf = df.groupby('DATE_HOUR').apply(compute_metrics)
```

```python
tf.shape
tf.to_csv('dataset/processed_SEA.csv',header=True)
```
